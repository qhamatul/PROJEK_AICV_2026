```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # Guna NumPy
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots() # Buat kanvas
ax.plot(x, y1, color='blue', label='Sin')
ax.plot(x, y2, color='red', label='Cos')

ax.set_title("Perbandinga Sin & Cos")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# data
nama_buah = ['Epal', 'Oren', 'Lemon']
jualan = [10, 15, 5]

# carta bar
plt.bar(nama_buah, jualan,color=['red', 'orange', 'yellow'])

# tajuk dan label
plt.title("Jualan Buah Hari Ini")
plt.xlabel("Nama Buah")
plt.ylabel("Bilangan Jualan")
plt.yticks(range(0, 16, 1))

# paparan graf
plt.show()
```
